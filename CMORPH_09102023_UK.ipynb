```python
from __future__ import print_function, division
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import numpy.ma as ma
from netCDF4 import Dataset
import cartopy.crs as crs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import matplotlib.cm as cm
import pandas as pd

def read_CMORPH_plot(ncf):
    nc_fid = Dataset(ncf, 'r')
    cmorph = nc_fid.variables["cmorph"][:]  # shape lat, lon as shown above
    lat = nc_fid.variables["lat"][:]
    lon = nc_fid.variables["lon"][:]
    names = nc_fid.variables.keys()
    nc_fid.close()
    return cmorph, lat, lon, names

cmorph_file = 'CMORPH2_0.25deg-30min_202309101500.RT.nc'
cmorph, lat, lon, names = read_CMORPH_plot(cmorph_file)
cmorph_1500 = cmorph[0,:,:]
cmorph_1500_w = cmorph[0,:,1400:1441]
cmorph_1500_e = cmorph[0,:,0:41]
cmorph_1500_con = np.concatenate((cmorph_1500_w, cmorph_1500_e), axis=1)
print("Concatenated Array: ", cmorph_1500_con, cmorph_1500_con.shape)
cmorph_1500_df = pd.DataFrame(cmorph_1500_con)
print(cmorph_1500_df)
cmorph_1500_df_int = cmorph_1500_df.ffill(axis=1)
print(cmorph_1500_df_int)
cmorph_1500_int = cmorph_1500_df_int.to_numpy()
print(cmorph_1500_int, cmorph_1500_int.shape)
lats,lons = np.meshgrid(lat, lon)
lats2 = np.swapaxes(lats, 0, 1)
lons2 = np.swapaxes(lons, 0, 1)
lats_w = lats2[:,1400:1441]
lats_e = lats2[:,0:41]
lats_con = np.concatenate((lats_w, lats_e), axis=1)
print("Concatenated Array: ", lats_con, lats_con.shape)
lons_w = lons2[:,1400:1441]
lons_w -= 360.0
lons_e = lons2[:,0:41]
lons_con = np.concatenate((lons_w, lons_e), axis=1)
print("Concatenated Array: ", lons_con, lons_con.shape)
print(names)
print('CMORPH:', cmorph_1500.shape, cmorph_1500)
print('Latitude:', lats2.shape, lats2)
print('Longitude:', lons2.shape, lons2)

del_cmorph_1500 = np.gradient(cmorph_1500_int)
del_cmorph_1500_x = del_cmorph_1500[0]
del_cmorph_1500_y = del_cmorph_1500[1]
print("Precip Gradient X = ", del_cmorph_1500_x)
print("Precip Gradient Y = ", del_cmorph_1500_y)
mag_1500 = np.sqrt(del_cmorph_1500[0]**2 + del_cmorph_1500[1]**2)
print("Precip Gradient = ", mag_1500)

cmorph_file = 'CMORPH2_0.25deg-30min_202309101530.RT.nc'
cmorph, lat, lon, names = read_CMORPH_plot(cmorph_file)
cmorph_1530 = cmorph[0,:,:]
cmorph_1530_w = cmorph[0,:,1400:1441]
cmorph_1530_e = cmorph[0,:,0:41]
cmorph_1530_con = np.concatenate((cmorph_1530_w, cmorph_1530_e), axis=1)
print("Concatenated Array: ", cmorph_1530_con, cmorph_1530_con.shape)
cmorph_1530_df = pd.DataFrame(cmorph_1530_con)
print(cmorph_1530_df)
cmorph_1530_df_int = cmorph_1530_df.ffill(axis=1)
print(cmorph_1530_df_int)
cmorph_1530_int = cmorph_1530_df_int.to_numpy()
print(cmorph_1530_int, cmorph_1530_int.shape)
lats,lons = np.meshgrid(lat, lon)
lats2 = np.swapaxes(lats, 0, 1)
lons2 = np.swapaxes(lons, 0, 1)
lats_w = lats2[:,1400:1441]
lats_e = lats2[:,0:41]
lats_con = np.concatenate((lats_w, lats_e), axis=1)
print("Concatenated Array: ", lats_con, lats_con.shape)
lons_w = lons2[:,1400:1441]
lons_w -= 360.0
lons_e = lons2[:,0:41]
lons_con = np.concatenate((lons_w, lons_e), axis=1)
print("Concatenated Array: ", lons_con, lons_con.shape)
print(names)
print('CMORPH:', cmorph_1530.shape, cmorph_1530)
print('Latitude:', lats2.shape, lats2)
print('Longitude:', lons2.shape, lons2)

del_cmorph_1530 = np.gradient(cmorph_1530_int)
del_cmorph_1530_x = del_cmorph_1530[0]
del_cmorph_1530_y = del_cmorph_1530[1]
print("Precip Gradient X = ", del_cmorph_1530_x)
print("Precip Gradient Y = ", del_cmorph_1530_y)
mag_1530 = np.sqrt(del_cmorph_1530[0]**2 + del_cmorph_1530[1]**2)
print("Precip Gradient = ", mag_1530)
```
